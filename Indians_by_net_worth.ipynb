```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Indians_by_net_worth"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.7559.134'}
 
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
tdata = soup.find_all(class_="wikitable sortable")
len(tdata)
```

```python
all_richest = tdata[0]
all_richest
```

```python
heads = [h.get_text().strip().replace('\n', ' ') for h in all_richest.find_all('th')]
heads
```

```python
data = all_richest.find_all('tr')[1:]
data
```

```python
rank, name, wc, nw, comp, sow = ([],[],[],[],[],[])

for d in data:
    res = d.find_all('td')
    rank.append(res[0].get_text().strip())
    name.append(res[1].get_text().strip())
    wc.append(res[2].find_all('span')[1]['title'])
    nw.append(res[3].get_text().strip())
    comp.append(res[4].get_text().strip())
    sow.append(res[5].get_text())
print(len(rank), len(name), len(wc), len(nw), len(comp), len(sow))
```

```python
df = pd.DataFrame({'Rank':rank, 'Name':name, 'Wealth Change':wc, 'Net Worth':nw, 'Company':comp,'Sources of Wealth':sow})
df
```

# International Indian Film Academy Awards

```python
url = "https://en.wikipedia.org/wiki/International_Indian_Film_Academy_Awards"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.7559.134'}
 
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
awards = soup.find(class_="sortable wikitable")
awards
```

```python
heads = [h.get_text().strip() for h in awards.find_all('th')]
heads
```

```python
tdata = awards.find_all('tr')[1:]
tdata
```

```python
cer, date, bfilm, hosts, venue, country = [],[],[],[],[],[]
for d in tdata:
    res = d.find_all('td')
    if len(res)==6:
        cer.append(res[0].get_text().strip())
        date.append(res[1].get_text().strip())
        bfilm.append(res[2].get_text().strip())
        host=res[3].get_text().strip()
        ven = res[4].get_text().strip()
        cou = res[5].get_text().strip()
        hosts.append(host)
        venue.append(ven)
        country.append(cou)
    elif len(res) == 5:
        cer.append(res[0].get_text().strip())
        date.append(res[1].get_text().strip())
        bfilm.append(res[2].get_text().strip())
        hosts.append(host)
        venue.append(res[3].get_text().strip())
        country.append(res[4].get_text().strip())
    else:
        cer.append(res[0].get_text().strip())
        date.append(res[1].get_text().strip())
        bfilm.append(res[2].get_text().strip())
        hosts.append(host)
        venue.append(ven)
        country.append(cou)
    
print(len(cer), len(date), len(bfilm), len(hosts), len(venue), len(country))
```

```python
q = [cer, date, bfilm, hosts, venue, country]
```

```python
list(zip(heads, q))
```

```python
d = dict(list(zip(heads, q)))
d
```

```python
df = pd.DataFrame(d)
df
```
